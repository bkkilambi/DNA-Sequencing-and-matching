```python
#Q3

def orf_codons(frame, dna_string):
    start_codon = ["ATG"]
    stop_codon = ["TAA", "TAG", "TGA"]
    #dna_string = list(dna_dict.values())[0]
    codon_list = []
    orf = []
    start_pos = -1
    stop_pos = -1

    for i in range(frame, len(dna_string), 3):
        codon_list.append(dna_string[i:i+3])
        
    for i in range(0,len(codon_list)):
        start_pos = -1
        stop_pos = -1
        if(codon_list[i] in start_codon):
            start_pos = i
            for j in range(i+1, len(codon_list)):
                if(codon_list[j] in stop_codon):
                    stop_pos = j
                    break
                else:
                    stop_pos = -1
            if(stop_pos != -1):
                orf.append(codon_list[start_pos:stop_pos+1])
        else:
            continue
    return orf

final_codon_orf = {}
## dictionary with id as key and list of lists which are ORF sequences for that value
for key, string in dna_dict.items():
    final_codon_orf[key] = orf_codons(1,string)
#orf_codons(0,'CTTAATATGTATATGTATCTATAGAAA')
#print(len(final_codon_orf['gi|142022655|gb|EQ086233.1|91']))
#dna_dict['gi|142022655|gb|EQ086233.1|91']
```
